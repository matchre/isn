Un point technique à voir tous ensemble : ouverture et lecture d'un fichier 
---------------------------------------------------------------------------

Pour ouvrir et lire un fichier contenant des données textuelles, on utilise une syntaxe du type `with open(chemin_du_fichier) as fichier:...` où les variables sont les suivantes.

1. `chemin_du_fichier` est une chaîne de caractères donnant le chemin d'accès à un fichier.

2. `fichier` est un nom de variable que vous pouvez changer.


Le fichier dont nous avons besoin dans cette section est un fichier texte contenant une liste des mots français.
Dans ce fichier, chaque ligne correspond exactement à un mot. La source utilisée se trouve dans [cette page](http://www.pallier.org/ressources/dicofr/dicofr.html).

En rédigeant ce document avec un sytème Unix, on se heurte à un problème technique classique.

### Un problème sous Windows aussi

Il ne faut pas croire que Windows a raison face aux systèmes Unix car le code suivant ne marchera pas comme attendu sous Windows avec les paramétrages par défaut.

```python
# Comme le fichier motsfrancais_frgut.txt se trouve dans le même dossier 
# que notre feuille IPython, nous n'avons pas besoin de donner son chemin
# absolu (on utilise donc un chemin relatif). 

# Nous n'allons afficher que les 5 premiers mots.
nb = 0

with open("motsfrancais_frgut_unix.txt") as fichier:
# Un fichier est une suite de lignes que l'on peut afficher les unes après
# les autres via une boucle `for` itérant directement sur la variable `fichier`.
    for ligne in fichier:
        nb += 1
        
        if nb <= 5:
# ATTENTION ! La méthode employée fait que la variable `ligne` contient 
# le caractère spéciale "\n" de retour à la ligne. Pour le supprimer,
# on utilise la méthode `strip`.  
            mot = ligne.strip()
            print(mot)

print('...')
```

### Que faire ? Quel est le problème ?

Soit on obtient un message d'erreur, soit le texte affiché n'est pas celui attendu. Expliquons d'où ceci vient. Par défaut les fichiers sont ouverts avec un certain encodage. Par exemple, sous Unix, c'est l'UTF-8. Python ne pouvant pas deviner à coup sûr un encodage non standard qu'utiliserait un fichier, il va lire le fichier avec l'encondage par défaut. 

Pour remédier à ce problème, il nous faut donc trouver le bon encodage et l'indiquer à Python afin d'obtenir un code qui marchera sans problème que l'on soit sous Windows ou sous un système Unix. On va se placer dans le cas d'un sytème Unix. 
Une astuce consiste à ouvrir un navigateur web pour ouvrir `motsfrancais_frgut.txt`, puis ensuite à changer l'encodage utilisé par le navigateur. Ceci permet de trouver que le bon encodage est `Occidental ISO Latin 1` bien aimé de Windows, mais pas des sytèmes Unix. Nous pouvons corriger notre code comme suit.

```python
# Nous n'allons afficher que les 20 premiers mots.
nb = 0

with open("motsfrancais_frgut.txt", encoding="iso-8859-1") as fichier:
    for ligne in fichier:
        nb += 1
        
        if nb <= 5:
            mot = ligne.strip()
            print(mot)

print('...')
```

Pour les plus rapides : une seconde liste de palindromes avec éventuellement des accents et/ou des traits d'union
-----------------------------------------------------------------------------------------------------------------

Si le problème précédent vous a paru simple, ou bien si vous l'avez bien intégré, vous allez essayer ici d'améliorer votre recherche de palindromes en ne tenant pas compte des accents et des traits d'union des mots composés. Par exemple, `réer` sera vu comme un palindrome.

Pour commencer, nous allons repérer tous les caractères spéciaux présents dans notre liste de mots. Ceci va être l'occasion de travailler avec un type de variables peu connu, mais très pratique, le type `set`, soit "ensemble" en français.

```python
# Au départ l'ensemble des lettres trouvées est l'ensemble vide, 
# c'est à dire l'unique ensemble ne contenant aucun élément.
lettres_trouvees = set()

with open("motsfrancais_frgut.txt", encoding="iso-8859-1") as fichier:
    for ligne in fichier:
        mot = ligne.strip()
        mot = mot.lower()

# Nous faisons la réunion de l'ensemble des lettres présentes dans
# la variable `mot` avec l'ensemble des lettres déjà trouvées.
        lettres_trouvees = lettres_trouvees | set(mot)

# Construisons l'ensemble des lettres de notre alphabet latin.
alphabet = set("abcdefghijklmnopqrstuvwxyz")

# Retirons de l'ensemble des lettres trouvées celles de notre
# alphabet.
lettres_speciales = lettres_trouvees - alphabet

# Créons à partir de l'ensemble `lettres_speciales` une liste ordonnée
# des lettres spéciales. On voit au passage que l'on peut transformer 
# un ensemble Python en une liste.
lettres_speciales_ordonnees = sorted(list(lettres_speciales))

# Affichons le texte obtenu en collant les différentes lettres spéciales
# séparées par une chaîne vide de caractères, autrement dit affichons le
# texte obtenu en joignant les différentes lettres spéciales. 
#
# ATTENTION ! la méthode `join` prend en variable une liste de chaînes
# de caractères.
print("".join(lettres_speciales_ordonnees))
```
